## User_phonenum 생성

```python
User_phonenum=[]

for _ in range(len(df1)):
    User_phonenum.append("010-" + str(time.time())[-4:] + "-" + str(time.time()).replace('.', "")[-5:-1])
#     time.sleep(0.001)

User_phonenum
```

## DATE 생성

```python
DATE=[]

date_ls=pd.date_range(start='2021-05-01', end='2021-06-02',periods=len(df1))
for i in date_ls:
    DATE.append(i.strftime("%y-%m-%d %H:%M:%S"))
    
DATE
```

## History 생성

## sales_history 생성
