```python
import pandas as pd
import re
```

```python
df=pd.read_csv('apriori_rules.csv',sep='/')
df
```

## item id -> 편의점 제품명 변경

```python
refer=pd.read_csv('prod_info_data.csv',index_col=0)[['name']]
refer
```

```python
mydic=refer.T.to_dict(orient='records')[0]
mydic
```

```python
df1=df.copy()
```

```python
for i in df1.rules:
    for j in re.findall('\d{5}', i):
        df1.rules=df1.rules.str.replace(j,mydic[int(j)])
df1
```

```python
df1.to_csv('apriori_rules_name.csv')
```
